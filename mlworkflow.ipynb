```python
# azureml/connect.py
# STEP 1: Connect to Azure ML workspace

# If you haven't already:
# !pip install azure-ai-ml azure-identity

from azure.identity import InteractiveBrowserCredential
from azure.ai.ml import MLClient

#  FILL THESE WITH YOUR REAL VALUES
SUBSCRIPTION_ID = "caa2ba3d-d055-4f23-8a4e-8684a42effdc"
RESOURCE_GROUP = "Machine_Learning_project_Marara"
WORKSPACE_NAME = "My-machine-learning-pro"

# This will use your Azure login (CLI, VS Code, browser, etc.)
credential = InteractiveBrowserCredential()

ml_client = MLClient(
    credential=credential,
    subscription_id=SUBSCRIPTION_ID,
    resource_group_name=RESOURCE_GROUP,
    workspace_name=WORKSPACE_NAME,
)

print(" Connected to workspace:", ml_client.workspace_name)

```

```python
local_movie_path = "https://raw.githubusercontent.com/Marara29/Machine-learning-engineering/refs/heads/main/enhanced_box_office_data(2000-2024)u.csv"
# Upload your file into workspaceblobstore/movies/

datastore_name = "workspaceblobstore"
target_path = "movies/enhanced_box_office_data.csv"  # blob path

ml_client.datastores.upload(
    name=datastore_name,
    src=local_movie_path,
    dest_path=target_path,
    overwrite=True,
)

print("Uploaded to blobstore at:", target_path)


```

```python
from azure.ai.ml.entities import Data
from azure.ai.ml.constants import AssetTypes

MOVIE_DATA_URL = "https://raw.githubusercontent.com/Marara29/Machine-learning-engineering/refs/heads/main/enhanced_box_office_data(2000-2024)u.csv"

DATA_ASSET_NAME = "movies-dataset"

movie_data_asset = Data(
    name=DATA_ASSET_NAME,
    description="Movies dataset from GitHub",
    path=MOVIE_DATA_URL,
    type=AssetTypes.URI_FILE,   # registering a single CSV file
)

registered = ml_client.data.create_or_update(movie_data_asset)

print(" Registered movies dataset:")
print("   Name   :", registered.name)
print("   Version:", registered.version)
print("   Path   :", registered.path)
```
